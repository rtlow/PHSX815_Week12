```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# just consider 10 samples per average
Nmeas = 10

for n in range(2, 7):
    
    # going to show how this converges with orders of magnitude
    Nexp = 10**n
    
    # sampling from a uniform distribution
    samples = np.random.rand(Nexp, Nmeas)
    
    # taking the mean
    means = np.mean(samples, axis=1)
    
    # fancy plotting stuff
    fig = plt.figure(figsize=[12,7])
    plt.hist(means, bins=100)
    plt.title('Average of 10^{} Experiments'.format(n))
    plt.xlabel('Average')
    plt.ylabel('Frequency')
    plt.show()
    plt.close(fig)
```

By the time we get to 10^6 experiments, we see that taking the average of 10 uniformly random samples produces a very Guassian distribution.
